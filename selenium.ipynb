```python
# you know pandas it's your best buddy 
import pandas as pd

# this library will be used for getting your credentials but keeping them hidden in the script 
from getpass import getpass

# these are the Classes we will use for interaction with a webpage
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# library for interacting with the operating system
import os

# this is the libraries that we will use to create break times in order to mimic human behaviour 
import time
import random
        

countries_capitals = pd.read_csv(".\capitals.csv")
cities = list(countries_capitals["capital"])
hotels_html = {}
```

```python
for city in cities:

    try:
        time.sleep(random.randint(2,5))

        # installing driver
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

        # opening wesite
        driver.get('https://www.hotelscombined.com/')
        time.sleep(random.randint(1,2))
        
        # clicking the searchbar
        search_bar = driver.find_element(By.CSS_SELECTOR, "div[class^='lNCO']")
        search_bar.click()

        time.sleep(random.randint(1,2))

        insert_section = driver.find_element(By.CSS_SELECTOR, "input[class^='k_my']")
        insert_section.click()

        time.sleep(random.randint(1,2))

        # inserting city
        insert_section.send_keys(city)
        insert_section.send_keys(Keys.ENTER)

        ### DATE --- HAVE MORE DATES TO TRY LATER 
        data = driver.switch_to.active_element
        data.send_keys(Keys.ENTER)

        time.sleep(random.randint(1,2))

        data = driver.switch_to.active_element
        data.send_keys(Keys.ENTER)

        time.sleep(random.randint(1,2))

        data = driver.switch_to.active_element
        data.send_keys(Keys.ENTER)

        ## Searching for HOTELS!!!!
        search_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label^='Search']")
        search_button
        search_button.click()

        time.sleep(random.randint(1,2))

        ## Scrolling
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.randint(8,10))

        ## Saving the HTML
        html = driver.page_source

        # Adding to the dictionary
        hotels_html[city]=html

        ## Closing Driver
        driver.close()

        # load pickle module
        import pickle

        # create a binary pickle file 
        f = open(city+".pkl","wb")

        # write the python object (dict) to pickle file
        pickle.dump(html,f)

        # close file
        f.close()

    except:
        ## Closing Driver
        driver.close()

        print (city, " not working")
```
