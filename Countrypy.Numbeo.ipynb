## Running with all capitals

```python
cities = list(countries_capitals['Capital'])
# cities = ['Lisbon','Madrid','Viena']

final = []
for city in cities:
    try:
        time.sleep(4)
        url = 'https://www.numbeo.com/cost-of-living/in/' + city

        params= {'displayCurrency' : 'EUR'}
        response = requests.get(url, params= params)

        response.status_code

        soup = BeautifulSoup(response.content)
        soup

        # avg prices in table - two frist rows:
        price_soup = soup.find_all('td', attrs= {'style':"text-align: right"})

        meal_prices_total = [price.text.replace("\xa0€","").strip() for price in price_soup]
        prices = meal_prices_total[0:2]
        prices

        #lable for prices - Two first rows
        lable_soup = soup.find_all('td')

        lable_soup = [lable.text.replace(",","").strip() for lable in lable_soup]
        lables = [lable_soup[2], lable_soup[5]]
        lables

        # Information together in a dictionary
        #dict = {lable : price for lable, price in zip(lables, prices)}
        dict = {'label' : lables[0], 'price' : prices[0], 'city' : city}
        
        final.append(dict)

        dict = {'label' : lables[1], 'price' : prices[1], 'city' : city}

        final.append(dict)

    except:
        print("This City doesn't work: ", city)
```

```python
final_df = pd.DataFrame(final)
final_df
```
